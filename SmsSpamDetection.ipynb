## Dataset is available here...
https://archive.ics.uci.edu/ml/datasets/SMS+Spam+Collection

```python
import pandas as pd
```

```python
messages = pd.read_csv('SMSSpamCollection', sep='\t', names=['label','message'])
```

```python
messages.head()
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
corpus=[]
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
# Creating the Bag of Words model
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500)
X = cv.fit_transform(corpus).toarray()
y = pd.get_dummies(messages['label'])
y=y.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

## Training model using Naive bayes classifier

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB().fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
```

```python
conf_matrix
```

```python
accuracy = accuracy_score(y_test, y_pred)
```

```python
accuracy
```

## Using Lemmatization and Tf-idf Vectorizer

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
corpus1 = []
for i in range(len(messages)):
    review = re.sub('^a-zA-Z', ' ', messages['message'][i])
    review = review.lower()
    review = review.split()
    review = [lemmatizer.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus1.append(review)  
```

```python
corpus1
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf = TfidfVectorizer(max_features=2500)
X1 = tf.fit_transform(corpus1).toarray()
```

```python
X1_train, X1_test, y_train, y_test = train_test_split(X1, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.naive_bayes import MultinomialNB
model1 = MultinomialNB().fit(X1_train, y_train)
```

```python
y_pred1 = model1.predict(X1_test)
```

```python
conf_matrix1 = confusion_matrix(y_test, y_pred1)
```

```python
conf_matrix1
```

```python
accuracy1 = accuracy_score(y_test, y_pred1)
```
